To detect performance degradation and identify potential issues affecting the prediction model.

```python
import pandas as pd
```

```python
import os
print(os.listdir("../../DATA"))
```

```python
new_data = pd.read_csv("../../DATA/simulated_new_data.csv.csv")

new_data.head()
```

Saving monitoring reports allows future comparison and supports model auditing.

To compare model predictions with the actual churn labels.

```python
X_new = new_data.drop(columns=["Churn Label"])

y_new = new_data["Churn Label"]
```

To evaluate the deployed model on the newly collected data.

```python
import joblib
```

```python
model = joblib.load("../../models/best_model.pkl")
```

To compare predicted values with actual customer churn.

```python
predictions = model.predict(X_new)

probabilities = model.predict_proba(X_new)[:,1]
```

To monitor whether the deployed model maintains acceptable performance.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

```python
accuracy = accuracy_score(y_new, predictions)

precision = precision_score(y_new, predictions)

recall = recall_score(y_new, predictions)

f1 = f1_score(y_new, predictions)

roc = roc_auc_score(y_new, probabilities)
```

To keep a monitoring history for future comparison.

```python
monitoring_data = pd.DataFrame({

    "Accuracy":[accuracy],

    "Precision":[precision],

    "Recall":[recall],

    "F1 Score":[f1],

    "ROC AUC":[roc]

})

monitoring_data
```

To maintain monitoring records for auditing and future analysis.

```python
monitoring_data.to_csv(

    "../../reports/model_monitoring.csv",

    index=False
)
```

To determine whether retraining is required.

```python
baseline_accuracy = 0.79

if accuracy < baseline_accuracy:

    print("⚠️ Performance Degradation Detected")

else:

    print("✅ Model Performance is Stable")
```
