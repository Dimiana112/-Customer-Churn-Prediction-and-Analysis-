```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("../../DATA/Telco_customer_churn.xlsx")
```

 Histograms
 Question

What is the distribution of numerical features?

```python
df.hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```

 Heatmap
 Question

Which numerical features are correlated?

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

Pair Plot
Question

How do numerical variables relate to each other?

```python
sns.pairplot(
    df.select_dtypes(include="number")
)

plt.show()
```

 Count Plot
 Question

How many customers churned and how many stayed?

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="Churn Label",
    data=df
)

plt.show()
```

 Count Plot
 Question

Does contract type affect customer churn?

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Contract",
    hue="Churn Label",
    data=df
)

plt.show()
```

 Count Plot
 Question

Does internet service influence customer churn?

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Internet Service",
    hue="Churn Label",
    data=df
)

plt.show()
```

 Box Plot
 Question

Are there outliers in Monthly Charges?

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x=df["Monthly Charges"]
)

plt.show()
```

 Box Plot
Question

Do churned customers have different monthly charges?

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x="Churn Label",
    y="Monthly Charges",
    data=df
)

plt.show()
```

# Key Insights

- Most customers did not churn.
- Month-to-month contracts have the highest churn rate.
- Customers with higher monthly charges tend to churn more.
- Total Charges is strongly correlated with Tenure Months.
- Most numerical variables show weak correlations.
- A few outliers exist in Monthly Charges and Total Charges.
