```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../../DATA/featured_telco.csv")
```

What is the overall churn distribution?

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Churn Label", data=df)

plt.title("Overall Customer Churn")

plt.show()
```

How does Monthly Charges differ between churned and non-churned customers?

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn Label",
    y="Monthly Charges",
    data=df
)

plt.title("Monthly Charges vs Churn")

plt.show()
```

Does customer tenure affect churn?

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn Label",
    y="Tenure Months",
    data=df
)

plt.title("Customer Tenure vs Churn")

plt.show()
```

Which contract type has the highest churn?

```python
contracts = pd.DataFrame()

contracts["Month-to-month"] = df["Contract_Month-to-month"]

contracts["One Year"] = df["Contract_One year"]

contracts["Two Year"] = df["Contract_Two year"]

contract_type = contracts.idxmax(axis=1)

sns.countplot(
    x=contract_type,
    hue=df["Churn Label"]
)

plt.xticks(rotation=15)

plt.title("Contract Type vs Churn")

plt.show()
```

Which payment method is associated with churn?

```python
payment = pd.DataFrame()

payment["Bank Transfer"] = df["Payment Method_Bank transfer (automatic)"]

payment["Credit Card"] = df["Payment Method_Credit card (automatic)"]

payment["Electronic Check"] = df["Payment Method_Electronic check"]

payment["Mailed Check"] = df["Payment Method_Mailed check"]

payment_type = payment.idxmax(axis=1)

sns.countplot(
    x=payment_type,
    hue=df["Churn Label"]
)

plt.xticks(rotation=20)

plt.title("Payment Method vs Churn")

plt.show()
```

Does internet service influence churn?

```python
internet = pd.DataFrame()

internet["DSL"] = df["Internet Service_DSL"]

internet["Fiber"] = df["Internet Service_Fiber optic"]

internet["No Internet"] = df["Internet Service_No"]

internet_type = internet.idxmax(axis=1)

sns.countplot(
    x=internet_type,
    hue=df["Churn Label"]
)

plt.title("Internet Service vs Churn")

plt.show()
```

Which numerical features are most correlated with churn?

```python
plt.figure(figsize=(12,8))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

Which features are the most important?

```python
selected_features = [
    "Tenure Months",
    "Online Security",
    "Monthly Charges",
    "Dependents",
    "Contract_One year",
    "Contract_Two year",
    "Internet Service_No",
    "Payment Method_Credit card (automatic)"
]

plt.figure(figsize=(8,5))

plt.barh(
    selected_features,
    range(len(selected_features))
)

plt.title("Selected Important Features")

plt.show()
```

Can we segment customers based on loyalty?

```python
sns.countplot(
    x="Loyal Customer",
    hue="Churn Label",
    data=df
)

plt.title("Loyal Customers vs Churn")

plt.show()
```

 Dashboard Insights
Most churned customers are on Month-to-Month contracts.
Customers with higher Monthly Charges are more likely to churn.
Customers with shorter Tenure show higher churn rates.
Fiber Optic users have a higher churn rate compared to DSL users.
Loyal customers generally have lower churn.
RFE identified Tenure, Monthly Charges, Contract Type, and Online Security as the most influential features for churn prediction.
