```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../DATA/processed_telco.csv")
```

```python
df.head()
```

Are there any opportunities to create new features?

```python
df.columns
```

Can we calculate the average monthly spending for each customer?

```python
df["Average Charges"] = (
    df["Total Charges"] /
    (df["Tenure Months"] + 1)
)
```

Does the new feature look reasonable?

```python
df["Average Charges"].describe()
```

Are all features already encoded?

```python
df.select_dtypes(include="object").columns
```

Are all numerical features scaled?

```python
df.describe()
```

Do we still have boolean columns?

```python
bool_columns = df.select_dtypes(include="bool").columns

bool_columns
```

Can we convert boolean features into integers?

```python
df[bool_columns] = df[bool_columns].astype(int)
```

```python
df.dtypes
```

Is the dataset ready for machine learning?

```python
df.info()
```

Can we estimate the customer's value per month?

```python
df["CLTV per Month"] = df["CLTV"] / (df["Tenure Months"] + 1)
```

```python
df[["CLTV","Tenure Months","CLTV per Month"]].head()
```

Can we identify high-paying customers?

```python
median_charge = df["Monthly Charges"].median()

df["High Monthly Charge"] = np.where(
    df["Monthly Charges"] > median_charge,
    1,
    0
)
```

```python
df["High Monthly Charge"].value_counts()
```

Can we identify loyal customers?

```python
median_tenure = df["Tenure Months"].median()

df["Loyal Customer"] = np.where(
    df["Tenure Months"] > median_tenure,
    1,
    0
)
```

```python
df["Loyal Customer"].value_counts()
```

How many new engineered features were created?

```python
new_features = [
    "Average Charges",
    "CLTV per Month",
    "High Monthly Charge",
    "Loyal Customer"
]

df[new_features].head()
```

Can we save the feature engineered dataset?

```python
df.to_csv("../../DATA/featured_telco.csv", index=False)
```
